```python
#=======================
#1-IMPORT DES LIBRAIRIES
#=======================
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns


#========================
#2-CHARGEMENT DES DONNEES
#========================
df = pd.read_csv('ventes.csv')
df['CA']=df['Prix']*df['Quantite']
print(df.head().to_string(index=False))


#=========================
#3-EXPLORATION DES DONNEES
#=========================
print('\n=== Statistiques descriptives ===')
print(df.describe())


#=====================
#4-ANALYSE DES DONNEES
#=====================
#Calcul du chiffre d'affaires total
ca_total = df['CA'].sum()
print(f'\nChiffre d\'affaires total : {ca_total:,}')

#Produit le plus vendu (en quantité)
produit_plus_vendu = df.groupby('Produit')['Quantite'].sum().sort_values(ascending=False).index[0]
print(f'\nProduit le plus vendu : {produit_plus_vendu}')

#Quantite vendue par produit
quantite_par_produit = df.groupby('Produit')['Quantite'].sum().sort_values(ascending=False).reset_index()
print('\nQuantite vendue par produit :')
print(quantite_par_produit.to_string(index=False))

#Calcul du chiffre d'affaires par categorie
ca_par_categorie = df.groupby('Categorie')['CA'].sum().sort_values(ascending=False).reset_index()
#print('\nChiffre d\'affaires par categorie :')
#print(ca_par_categorie.to_string(index=False))

#calcul du chiffre d'affaires par produit
ca_par_produit = df.groupby('Produit')['CA'].sum().sort_values(ascending=False).reset_index()
#print('\nChiffre d\'affaires par produit :')
#print(ca_par_produit.to_string(index=False))

#======================
#5-VISUALISATION DES DONNEES
#======================

#Bar plot du chiffre d'affaires par categorie
sns.set(style='whitegrid',palette='pastel')
plt.figure(figsize=(10,6))
sns.barplot(x='Categorie', y='CA',data=ca_par_categorie,palette='viridis')
plt.title('Chiffre d\'affaires par catégorie')
plt.xlabel('Catégorie')
plt.ylabel('Chiffre d\'affaires')
plt.tight_layout()
plt.savefig('categorie.png')
plt.show()
 
 #Bar plot du chiffre d'affaires par produit
plt.figure(figsize=(12,6))
sns.barplot(x='Produit', y='CA',data=ca_par_produit,palette='magma')
plt.title('Chiffre d\'affaires par produit')
plt.xlabel('Produit')
plt.ylabel('Chiffre d\'affaires')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('produit.png')
plt.show()

#Graphique de la quantité des ventes par produit
plt.figure(figsize=(12,6))
sns.barplot(x='Quantite', y='Produit', data=quantite_par_produit,palette='coolwarm')
plt.title('Quantité des ventes par produit')
plt.xlabel('Quantité vendue')
plt.ylabel('Produit')
plt.tight_layout()
plt.savefig('distribution.png')
plt.show()
```
